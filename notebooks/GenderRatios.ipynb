```python
df_ratios_melt = df_ratios.reset_index().melt(id_vars='id', var_name='ratio_name', value_name='ratio_value')
df_ratios_melt_rat = df_ratios_melt[df_ratios_melt.ratio_name.str.startswith('ratio_')]
df_ratios_melt_num = df_ratios_melt[df_ratios_melt.ratio_name.str.startswith('num_')]
df_ratios_melt
```

```python
df_meta = get_corpus_metadata_annotated()
df = df_ratios_melt_rat.merge(df_meta, on='id')
df['imdb_release_date'] = df['imdb_release_date'].apply(float)
df

import plotnine as p9
p9.options.figure_size = 11,8
fig = p9.ggplot(df.query('imdb_release_date>1975 & 0.1<ratio_value<100'), p9.aes(x='imdb_release_date', y='ratio_value'))
fig+= p9.geom_point(size=.1)
fig+= p9.ylim(0,10)
fig+= p9.scale_y_log10()
fig+= p9.facet_wrap('ratio_name')
fig+= p9.geom_smooth()
fig+= p9.geom_hline(yintercept=1, size=.25)
fig
```
