```python
def get_subcorpus():
    url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTTnPHG0eNrPcuR1pNWrGEy05LLFBf0Lpty6SEeinTn0Mq5bFsNg592qVPxXzZPSgTWjtK944en-yaD/pub?gid=572829200&single=true&output=csv'
    df = pd.read_csv(url, dtype='str').fillna('').drop_duplicates('id')
    df['subgenre'].value_counts()
    df['is_superhero'] = df.subgenre.apply(lambda g: g.lower().strip() in {'marvel', 'dceu', 'superhero'})
    dfsub = df[df.is_superhero].sort_values(['tmdb_release_date','id'])
    dfsub = dfsub[dfsub['correct_imdb']!='']
    return dfsub
```

```python
# !pip install cinemagoer

def get_imdb_meta(
        idx,
        keys=['rating', 'votes', 'imdbID', 'title', 'year']):
    idx=str(idx)
    if idx.startswith('tt'): idx=idx[2:]
    while len(idx)<7: idx='0'+idx

    from imdb import Cinemagoer
    ia = Cinemagoer()
    movie = ia.get_movie(idx)

    imdb_data = {k:movie[k] for k in keys}
    imdb_data['correct_imdb']='tt'+idx
    return imdb_data
```

```python
def get_subcorpus_castlist(dfsub = None):
    if dfsub is None: dfsub = get_subcorpus()
    o = []
    def fixname(x):
        while '  ' in x: x=x.replace('  ',' ')
        return x

    for idx in tqdm(dfsub.id):
        dfcast = get_text_cast(idx)#[cols]
        dfcast['actor_name'] = dfcast['actor_name'].apply(fixname)
        dfcast['char_name'] = dfcast['char_name'].apply(fixname)
        dfcast['actor_gender'] = [x.title()[0] if y and float(y)>=.9 else '?' for x,y in zip(dfcast.actor_gender, dfcast.actor_probability)]
        dfcast['char_gender'] = [x.title()[0] if y and float(y)>=.9 else '?' for x,y in zip(dfcast.char_gender, dfcast.char_probability)]
        # dfcast['actor_gender_anno'] = ''
        # dfcast['char_gender_anno'] = ''
        dfcast['cast_order'] = [i+1 for i in range(len(dfcast))]
        dfcast['movie_id'] = idx
        o.append(dfcast[['movie_id', 'cast_order', 'actor_name', 'actor_gender', 'char_name', 'char_gender']])
        # break
    odf = pd.concat(o).set_index(['movie_id','cast_order']).sort_index()
    odf = odf[~odf.char_name.str.contains('(uncredited)')]
    return odf
```
