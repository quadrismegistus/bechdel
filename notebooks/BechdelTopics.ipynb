```python
def get_speeches_by(self,by='actor_gender', on='char_name', hegemonic_category='CM'):
    dfcast = self.get_cast()
    is_hegemon = dict(zip(dfcast[on], dfcast[by].apply(lambda x: x==hegemonic_category)))
    counters = {}
    for intd in self.iter_interactions():
        spkr=intd.get('source')
        rcvr=intd.get('target')
        if spkr in is_hegemon and rcvr in is_hegemon:
            if not is_hegemon[spkr] and not is_hegemon[rcvr]:
                int_type = f'Speech_NOT_Involving_{hegemonic_category}'
            else:
                int_type = f'Speech_Involving_{hegemonic_category}'
            
            if not int_type in counters: counters[int_type]=[]
            counters[int_type]+=[intd['speech']]
    return counters
```
