```python
# import
import sys; sys.path.insert(0,'..')
from bechdeltest import *
PATH_MOVIESCRIPT_ORIG = '/Users/ryan/data/moviedb/Movie-Script-Database/scripts/'
```

```python
def postproc_moviescriptdb_metadata():
    # Read metadata
    meta_fn=os.path.join(PATH_MOVIESCRIPT_ORIG,'metadata','clean_parsed_meta.json')
    with open(meta_fn) as f: data = json.load(f)

    def parse_rowd(rowd):
        odx = {}
        for keytype in rowd:
            for key,val in rowd[keytype].items():
                #odx[f'{keytype}_{key}'] = val
                odx[key] = val
        return odx

    index = data.keys()
    rows = [parse_rowd(data[idx]) for idx in index]
    df = pd.DataFrame(rows, index=index).rename_axis('id_orig')
    df['id'] = df.file_name
    df = df.set_index('id').drop('dialogue	charinfo	tagged name'.split(), axis=1)
    df = df[['title','release_date','source', 'script_url', 'file_name', 'overview']]
    df.to_csv(PATH_CORPUS_METADATA)
    return df
```

```python
df = postproc_moviescriptdb_metadata()
df
```

## Postproc files

```python
def postproc_moviescriptdb_files():
    df['path_csv'] = df.file_name.apply(
        lambda x: os.path.join(PATH_MOVIESCRIPT_ORIG,'parsed_movie_scripts', x+'.csv')
    )
    objs = list(df.path_csv)#[:2]
    pmap(do_postproc_moviescriptdb_files, objs, num_proc=1)


def do_postproc_moviescriptdb_files(fn):
    if not os.path.exists(fn): return
    fnid=os.path.splitext(os.path.basename(fn))[0]
    odir=os.path.join(PATH_CORPUS_TEXTS,fnid)
    if not os.path.exists(odir): os.makedirs(odir)
    ofn = os.path.join(odir, f'{fnid}.dialogue.csv')
    shutil.copy(fn, ofn)
```

```python
postproc_moviescriptdb_files()
```
