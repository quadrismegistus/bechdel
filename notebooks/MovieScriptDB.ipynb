```python
fn='/Users/ryan/data/moviedb/Movie-Script-Database/scripts/metadata/clean_parsed_meta.json'
```

```python
import json,pandas as pd
```

```python
with open(fn) as f: data = json.load(f)
```

```python
def parse_rowd(rowd):
    odx = {}
    for keytype in rowd:
        for key,val in rowd[keytype].items():
            odx[f'{keytype}_{key}'] = val
    return odx

parse_rowd(data['10thingsihateaboutyou'])
```

```python
index = data.keys()
rows = [parse_rowd(data[idx]) for idx in index]
df = pd.DataFrame(rows, index=index).rename_axis('id')
df
```
